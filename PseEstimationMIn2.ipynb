```python
#import modules
import cv2
import mediapipe as mp
import time

#create model
mpDraw = mp.solutions.drawing_utils #For drawing key points
mpPose = mp.solutions.pose # Initializing mediapipe pose class.
pose = mpPose.Pose()
cap = cv2.VideoCapture('media/2.mp4') #read files for specified path
pTime = 0
while True:
    success, img = cap.read() #read files
    img = cv2.resize(img, dsize=(1000,700)) #resize video
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #converting the image into RGB format
    results = pose.process(imgRGB) # Perform pose detection
    
    if results.pose_landmarks: # Check if any landmarks are found
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS) # Draw Pose landmarks on the output points and lines
        for id, lm in enumerate(results.pose_landmarks.landmark): 
            h, w, c = img.shape #height, width and channel
            print(id, lm) #print landmarks
            cx, cy = int(lm.x * w), int(lm.y * h)
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)#big circle on out
    if results.pose_landmarks:
        for i in range(32):
        
        # Display the found normalized landmarks.
            print(f'{mpPose.PoseLandmark(i).name}:\n{results.pose_landmarks.landmark[mpPose.PoseLandmark(i).value]}') 
#adjust frame rate per second    
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3,
                (255, 0, 0), 3)

    cv2.imshow("Image", img)
    cv2.waitKey(1) #wait 1 millisecond
```
